```python
import os, zipfile, urllib.request

url = "https://github.com/Monirekhonsari/analogy_using_embeddings/releases/download/v1.0.0/glove.6B.zip"
zip_path, _ = urllib.request.urlretrieve(url, "glove.6B.zip")

with zipfile.ZipFile(zip_path) as z:
    z.extractall("glove_files")

print(os.listdir("glove_files"))
```

```python
import numpy as np
glove_dir = 'glove_files'

embeddings_index = {}

with open(os.path.join(glove_dir, 'glove.6B.100d.txt'), encoding='utf8') as f:
    for line in f:
        values = line.strip().split()
        word, nums = values[0], values[1:]        #شروع توضیح
        coefs = np.asarray(nums, dtype='float32')
        embeddings_index[word] = coefs
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similarity(u, v):
  return np.squeeze(cosine_similarity(u.reshape(1, -1), v.reshape(1, -1)))
```

Returns the cosine similarity of two vectors (a number between -1 and 1)

```python
father = embeddings_index['father']
mother = embeddings_index['mather']
ball = embeddings_index['ball']
crocodile = embeddings_index['crocodile']
france = embeddings_index['france']
paris = embeddings_index['paris']
iran = embeddings_index['iran']
tehran = embeddings_index['tehran']

print('cosine-similarity(father, mother) = ', similarity(father, mother))
print("cosine_similarity(ball, crocodile) = ",similarity(ball, crocodile))
print("cosine_similarity(france - paris, tehran - iran) = ",similarity(france - paris, tehran - iran))
print("cosine_similarity(paris - france, iran - tehran ) = ",similarity(paris - france, iran - tehran))
```

```python
def complete_analogy(word_a, word_b, word_c, embeddings_index):

    word_a, word_b, word_c = word_a.lower(), word_b.lower(), word_c.lower()

    e_a, e_b, e_c = embeddings_index[word_a], embeddings_index[word_b], embeddings_index[word_c]

    words = embeddings_index.keys()
    max_cosine_sim = -100              # Initialize max_cosine_sim to a large negative number
    best_word = None                   # Initialize best_word with None, it will help keep track of the word to output

    # loop over the whole word vector set
    for w in words:
        # to avoid best_word being one of the input words, pass on them.
        if w in [word_a, word_b, word_c] :
            continue

        # Compute cosine similarity between the vector (e_b - e_a) and the vector ((w's vector representation) - e_c)  (≈1 line)
        cosine_sim = similarity(e_b - e_a, embeddings_index[w] - e_c)

        # If the cosine_sim is more than the max_cosine_sim seen so far,
            # then: set the new max_cosine_sim to the current cosine_sim and the best_word to the current word (≈3 lines)
        if cosine_sim > max_cosine_sim:
            max_cosine_sim = cosine_sim
            best_word = w

    return best_word
```
